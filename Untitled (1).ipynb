```python
import yfinance as yf
import pandas as pd

# Crear objeto Ticker para Tesla
tesla = yf.Ticker("TSLA")

# Descargar datos históricos de Tesla
tesla_data = tesla.history(period="max")

# Restablecer el índice
tesla_data.reset_index(inplace=True)

# Guardar los datos en un archivo CSV
tesla_data.to_csv("tesla_data.csv", index=False)

# Mostrar las primeras cinco filas
print(tesla_data.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL con los datos de ingresos de Tesla
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# Realizar la solicitud HTTP con encabezados para evitar bloqueos
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
response = requests.get(url, headers=headers)

# Verificar que la solicitud sea exitosa
if response.status_code == 200:
    print("Conexión exitosa.")
else:
    print(f"Error al conectar: {response.status_code}")

# Analizar el contenido HTML con BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Buscar la tabla de ingresos (normalmente es la segunda tabla)
table = soup.find_all("table")[1]  # Cambia el índice si la tabla no es la segunda

# Extraer las filas de la tabla
rows = table.find_all("tr")

# Procesar los datos en una lista
tesla_revenue = []
for row in rows[1:]:
    cols = row.find_all("td")
    tesla_revenue.append([col.text.strip() for col in cols])

# Crear un DataFrame con los datos extraídos
tesla_revenue_df = pd.DataFrame(tesla_revenue, columns=["Date", "Revenue"])

# Limpiar los datos si es necesario (por ejemplo, eliminar filas vacías)
tesla_revenue_df = tesla_revenue_df.dropna()

# Mostrar las últimas cinco filas
print(tesla_revenue_df.tail())

# Guardar los datos en un archivo CSV
tesla_revenue_df.to_csv("tesla_revenue.csv", index=False)
```

```python
import yfinance as yf
import pandas as pd

# Descargar los datos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Restablecer el índice y guardar en un archivo CSV
tesla_data.reset_index(inplace=True)
tesla_data.to_csv("tesla_data.csv", index=False)

# Asegúrate de que la columna 'Date' sea un tipo datetime
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])
```

```python
import matplotlib.pyplot as plt

# Función para graficar datos de Tesla
def make_graph(data, title):
    """
    Crea un gráfico de los precios de cierre de acciones.
    
    Parámetros:
    - data: DataFrame con columnas 'Date' y 'Close'.
    - title: Título del gráfico.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label="Precio de Cierre", color="blue")
    plt.title(title, fontsize=16)
    plt.xlabel("Fecha", fontsize=14)
    plt.ylabel("Precio de Cierre (USD)", fontsize=14)
    plt.grid(True)
    plt.legend(fontsize=12)
    plt.show()
```

```python
# Cargar los datos de Tesla desde un archivo CSV si ya está guardado
tesla_data = pd.read_csv("tesla_data.csv")
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])

# Llamar a la función para graficar
make_graph(tesla_data, "Gráfico de Precios de Cierre de Tesla")
```
